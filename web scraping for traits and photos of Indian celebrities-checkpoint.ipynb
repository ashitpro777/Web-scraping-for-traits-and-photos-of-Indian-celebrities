```python
from bs4 import BeautifulSoup
import requests
import csv
import shutil
#Website to scrap Zodiac of some popular Indian celebrities
source= requests.get('https://www.boldsky.com/astrology/zodiac-signs/indian-celebrities-and-their-zodiac-signs/articlecontent-pf196406-120860.html').text
#Website to scrap Traits of the those Indian celebrities 
source2= requests.get('https://www.letuspublish.com/bollywood-celebrities-and-their-zodiac-sign/').text
#Website to scrap Photos of those Indian celebrities(MALE)
source3 = requests.get('https://www.imdb.com/list/ls025929404/?sort=list_order,asc&mode=detail').text
#Website to scrap Photos of those Indian celebrities(FEMALE)
source4 = requests.get('https://www.imdb.com/list/ls025928763/').text

soup = BeautifulSoup(source,'lxml')
soup2 = BeautifulSoup(source2,'lxml')
soup3i = BeautifulSoup(source3,'lxml')
soup3ii = BeautifulSoup(source4,'lxml')

match = soup.find('article')
#print(match.prettify())

tempfile1=open(r'Temp_file1.csv','w')
tempwriter=csv.writer(tempfile1)
tempwriter.writerow(['Photo URL','Celebrities Name','Zodiac','Traits'])

for x in match.find_all('div',class_='listicalSliderContainer'):
    zodiac=x.h2.text.strip()
    trait=x.p.text
    try:
        y=x.em.text.split(':')[1]
    except Exception as e:
        y=x.em.text.split('are')[1]
    name1=y.rsplit(' and ')[1].split('.')[0]
    name2=y.rsplit(' and ')[0].split(',')
    names=[name1,*name2]
    #print(names)
    for z in names:
        #print(z,zodiac,trait)
        tempwriter.writerow(["None",z.strip(),zodiac,trait])
tempfile1.close()


tempfile2=open(r'Temp_file2.csv','w')
tempwriter2=csv.DictWriter(tempfile2,['Photo URL','Celebrities Name','Zodiac','Traits'])

match2=soup2.find('article')
#print(match2.prettify())

m=match2.find('div',class_='td-post-content')

for v in m.find_all('ol'):
    zodiac2=v.text.split(':')[0].strip()
    trait2=v.next_sibling.next_sibling.next_sibling.text.strip()   
    with open('Temp_file1.csv','r') as csvfile:
        reader=csv.DictReader(csvfile)
        for row in reader:
            if row['Zodiac'] == zodiac2:
                row['Traits']=trait2
                row={'Photo URL':row['Photo URL'],'Celebrities Name':row['Celebrities Name'],'Zodiac':row['Zodiac'],'Traits':trait2}
               # print(row)
                tempwriter2.writerow(row)
    csvfile.close()
tempfile2.close()

finalfile=open('Final_csv_file.csv','w')
tempwriter3=csv.DictWriter(finalfile,['Photo URL','Celebrities Name','Zodiac','Traits'])

match3 = soup3i.find('div',class_='lister list detail sub-list')
match4 = soup3ii.find('div',class_='lister list detail sub-list')
#print(match3.prettify())
#print(match4.prettify())

with open('Temp_file2.csv','r') as csv2file:
    reader=csv.DictReader(csv2file,['Photo URL','Celebrities Name','Zodiac','Traits'])
    for row in reader:
        j="None"
        for i in match3.find_all('div',class_='lister-item mode-detail'):
            k=i.h3.a.text.strip()
            if row['Celebrities Name'] == k:
                #print("@")
                j=i.find('div',class_='lister-item-image').a.img.get('src')
        if j == "None":
            for n in match4.find_all('div',class_='lister-item mode-detail'):
                m=n.h3.a.text.strip()
                if row['Celebrities Name'] == m:
                    j=n.find('div',class_='lister-item-image').a.img.get('src')
        row={'Photo URL':j,'Celebrities Name':row['Celebrities Name'],'Zodiac':row['Zodiac'],'Traits':row['Traits']}
        #print(j)
        print(row)
        tempwriter3.writerow(row)
    csvfile.close()
finalfile.close()
                
```
